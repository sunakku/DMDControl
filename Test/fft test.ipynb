```python
def imload(path):
    return np.array(pil.open(path))[:,:,1]/255.
def showimage(image,title):
    plt.imshow(image)
    plt.gray()
    plt.title(title)
    plt.colorbar()
```

# np.fft.fft2

```python
def numpyfft(im):
    fft = np.fft.fft2(im)
    plt.figure(1,(15,5))
    plt.subplot(1,3,1)
    showimage(fft.real,"real")
    plt.subplot(1,3,2)
    showimage(np.absolute(fft),"abs")
    plt.subplot(1,3,3)
    showimage(np.angle(fft),"phase")
    plt.show()
```

```python
fft = np.fft.fft2(im)
plt.figure(1,(15,5))
plt.subplot(1,3,1)
showimage(np.log(fft.real),"real log")
plt.subplot(1,3,2)
showimage(np.log(np.absolute(fft)),"abs log")
plt.subplot(1,3,3)
showimage(np.angle(fft)[30:40,30:40],"phase")
plt.show()
```
